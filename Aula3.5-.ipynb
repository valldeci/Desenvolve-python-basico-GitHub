```python
#Q1
#calculadora
n1 = float(input())
operador = input()
n2 = float(input())

if operador == "+":
    resultado = n1 + n2
elif operador == "-":
    resultado = n1 - n2
elif operador == "/":
    resultado = n1 / n2
elif operador == "*":
    resultado = n1 * n2

print(f"{n1}{operador}{n2}={resultado}")

#Q2
#triangulo simpes
l1 = input('Lado 1: ')
l2 = input('Lado 2: ')
l3 = input('Lado 3: ')

if l1 == l2 or l2 == l3 or l1 == l3:
    print('Isósceles')
elif l1 == l2 and l2 == l3:
    print('Equilátero')
else:
    print('Escaleno')
    
#Q3
#pontuação 
pontuacao = int(input("Digite a pontuação do jogador: "))

if pontuacao >= 90:
    classificacao = "Excelente"
elif pontuacao >= 80:
    classificacao = "Bom"
elif pontuacao >= 70:
    classificacao = "Regular"
else:
    classificacao = "Insatisfatório"

print("Classificação:", classificacao)


#Q4

valor = float(input("Digite o valor total da compra: "))

if valor >= 100:
    desconto = 20.0
elif valor >= 50:
    desconto = 10.0
else:
    desconto = 0.0

valor_final = valor - (valor * desconto / 100)

print(f"Desconto aplicado: {desconto}%")
print(f"Valor final com desconto: R${valor_final:.2f}")


#Q5 
# Banco de dados 
usuarios = {
    "admin": ("admin123", "Administrador"),
    "usuario1": ("senha123", "Usuário Comum"),
    "visitante1": ("guest2024", "Visitante")
}

while True:
    print("\n=== Sistema de Autenticação ===")
    login = input("Usuário (ou digite 'sair' para encerrar): ")

    if login.lower() == "sair":
        print("Encerrando o sistema...")
        break

    senha = input("Senha: ")

    # Verifica se o usuário existe no banco de dados
    if login in usuarios:
        senha_correta, nivel = usuarios[login]

        # Verifica se a senha está correta
        if senha == senha_correta:
            print("\nLogin bem-sucedido! ✅")
            print(f"Bem-vindo, {login}.")
            print(f"Nível de acesso: {nivel}")
        else:
            print("Erro: Usuário ou senha inválidos.")
    else:
        print("Erro: Usuário ou senha inválidos.")


    
```
